## Study of missing values
As we can see from the above table, our data has missing values for the "Age", "Cabin", and "Embarked" fields.

We must assess the proportion of the missing data in the train set to help us decide how to deal with them.

```julia
nmissing_age = 0
missing_age_index = []
# iterate through each row
for index in 1:size(data)[1]
    # check if age is missing
    if ismissing(data[index, :Age])
        nmissing_age += 1
        push!(missing_age_index, index)
    end
end
print(nmissing_age)
```

```julia
nsurvived_missing_age = 0
for value in eachrow(data[missing_age_index, :])
    if value.Survived == 1
        nsurvived_missing_age += 1
    end
end
print(nsurvived_missing_age)
```

### A large part of the data has missing values for Age, this can turn out to be problematic for the ML algorithms...
A possible solution would be to drop the "Age" column if there is not a big correlation with target labels.

Another would be to replace missing values with random integers sampled according to the mean and variance of the available data.

```julia
nmissing_cabin = 0
for index in 1:size(data)[1]
    # check if cabin is missing
    if ismissing(data[index, :Cabin])
        nmissing_cabin += 1
    end
end
print(nmissing_cabin)
```

### The majority of our data does not have a value for the "Cabin" field.
This doesn't leave much to choose from..

Again, we can simply drop the field.

Or map it to a new "HasCabin" property.

```julia
data[:, [:Cabin, :Pclass]]
```

### We can notice that it is more common for people in second or third class to have a missing cabin.
Let's study the relation between these two properties

```julia
m = zeros(Int64, 3, 2)

for value in eachrow(data)
    if value.Pclass == 1
        if ismissing(value.Cabin)
            m[1, 2] = m[1, 2] + 1
        else
            m[1, 1] = m[1, 1] + 1
        end
    elseif value.Pclass == 2
        if ismissing(value.Cabin)
            m[2, 2] = m[2, 2] + 1
        else
            m[2, 1] = m[2, 1] + 1
        end
    else
        if ismissing(value.Cabin)
            m[3, 2] = m[3, 2] + 1
        else
            m[3, 1] = m[3, 1] + 1
        end
    end
end
```

```julia
groupedbar(m, bar_position = :stack, 
    title="Proportion of present (blue) \nand missing (red) cabin value per class", 
    xlabel="Class", ylabel="Number of individuals", legend=false)
```

### It seems that our guess is correct.
Because there is a high correlation between the pressence of a cabin and the class of person, we can choose to discard the "Cabin" property.

### Name
It is somewhat hard to extract anything useful from the name field but after some research (source: https://triangleinequality.wordpress.com/2013/09/08/basic-feature-engineering-with-the-titanic-data/), there is title present (eg: Mr, Miss, Master etc) in each name. 

We could use this title to infer other information such as sex (Mr/Mrs), age (Master/Miss) or status (Don, Rev, Capt etc).

```julia
# function to map name to title eg: mr, miss, master, doctor etc
function substring_in_string(big_string, substrings)
    for substring in substrings
        if occursin(substring, big_string)
            return substring
        end
    end
    return missing
end

titles_list = ["Mrs", "Mr", "Master", "Miss", "Major", "Rev",
                    "Dr", "Ms", "Mlle","Col", "Capt", "Mme", "Countess",
                    "Don", "Jonkheer"];
```

```julia
data.Title = map(x -> substring_in_string(x, titles_list), data.Name);
```

```julia
m = zeros(Int64, size(titles_list)[1], 2)
nam = titles_list

for value in eachrow(data)
    if value.Survived == 0
        m[findfirst(isequal(value.Title), titles_list), 1] += 1
    else
        m[findfirst(isequal(value.Title), titles_list), 2] += 1
    end
end
```

```julia
groupedbar(nam, m, bar_position = :stack, 
    title="Survivability per title \nSurvived (orange), Died (blue)", 
    xlabel="title", ylabel="Number of individuals", legend=false)
```

```julia
for title in titles_list
    all_with_title = filter(x -> x.Title == title, eachrow(data))
    nwith_title_survived = count(x -> x.Survived == 1, all_with_title)
    perc = nwith_title_survived / size(all_with_title)[1]
    println("Survival rate for $(title): $perc")
end
```

```julia
count(x -> x.Title == "Dr", eachrow(data))
```

As expected, titles that correspond to members of the fair sex have a higher chance to survive.

On the other hand, people that have the title "Master" (with 40 examples) has a disproportionatly high chance of survival for their sex.

The other titles are too scarce (1 to 7 occurrences) to allow us to draw any meaningful conclusions.

```julia
titles = map(x -> findfirst(isequal(x.Title), titles_list), eachrow(data))
cor(data.Survived, titles)
```

```julia
cor(sex, titles)
```

### There is some correlation between title and survability
but we can safely blame it on its relation to sex.

For training the classification algorithms, we could easily choose to ignore the "Title" and "Name" properties but taking into consideration the "Master" title could prove to be handy.

### Family study

```julia
family_dict = Dict()
```

```julia
for value in eachrow(data)
    # regex to extract first name
    regex = r"^([\w\-]+)"
    # extract string
    name = match(regex, value.Name).match
    if haskey(family_dict, name)
        # push survived value to set matching the name 
        push!(get(family_dict, name, nothing), value.Survived)
    else
        # if dict does not contain this family, create a new set containing the "Survived" value of this person
        family_dict[name] = [value.Survived]
    end
end
```

```julia
# array stores:
# number of families for which all members dies in index 1
# ".." members die and survive in index 2
# ".." all members survive in index 3
m = zeros(Int64, 3)

fam_sum = 0
fam_num = 0

for (key, value) in family_dict
    # family has more than 1 member
    if size(value)[1] > 1
        fam_sum += size(value)[1]
        fam_num += 1
        if 0 ∈ value
            if 1 ∈ value
                # mixed survival
                m[2] += 1
            end
            # all member die
            m[1] += 1
        else
            # all members survive
            m[3] += 1
        end
    end
end
```

```julia
bar(["All die", "Mixed survival", "All survive"], m,
    title="Survival per members of family", ylabel="Number of families", legend=false)
```

```julia
fam_sum / fam_num
```

### Let us try to create some basic classifiers with the paterns found so far and compare their performace.  

```julia
function predict_all_die(example)
    return 0
end
```

```julia
function predict_all_women_surive(example)
    if example.Sex == "female" #|| (!ismissing(example.IsYoung) && example.IsYoung == 1) || example.Pclass == 1
        return 1
    else
        return 0
    end
end
```

```julia
function predict_all_women_and_children_survive(example)
    if example.Sex == "female" || (!ismissing(example.IsYoung) && example.IsYoung == 1)# || example.Pclass == 1
        return 1
    else
        return 0
    end
end
```

```julia
function predict_all_women_survive_using_family(example, family_dict)
    regex = r"^([\w\-]+)"
    name = match(regex, example.Name).match
    if example.Sex == "female"
        if haskey(family_dict, name)
            value = get(family_dict, name, missing)
            if size(value)[1] > 1 && mean(value) == 0
                return 0
            end
        end
        return 1    
    else
        if haskey(family_dict, name)
            value = get(family_dict, name, missing)
            if size(value)[1] > 1 && mean(value) == 1
                return 1
            end
        end
        return 0
    end
end
```

```julia
function predict_all_women_and_1class_survive(example)
    if example.Sex == "female" || example.Pclass == 1
        return 1
    else
        return 0
    end
end
```

```julia
function predict_all_children_and_1class_survive(example)
    if (!ismissing(example.IsYoung) && example.IsYoung == 1) || example.Pclass == 1
        return 1
    else
        return 0
    end
end
```

```julia
function score(predicted, actual)
    correct = 0
    for i in 1:size(predicted)[1]
        if predicted[i] == actual[i]
            correct += 1
        end
    end
    return correct / size(predicted)[1]
end
```

```julia
function get_prediction(data, func, family_dict=missing)
    ŷ = []
    if ismissing(family_dict)
        for value in eachrow(data)
            push!(ŷ, func(value))
        end
    else
        for value in eachrow(data)
            push!(ŷ, func(value, family_dict))
        end
    end
    return ŷ
end
```

```julia
function get_family_dict(data)
    family_dict = Dict()
    
    # regex to extract first name
    regex = r"^([\w\-]+)"
    # extract string
    for value in eachrow(data)
        name = match(regex, value.Name).match
        if haskey(family_dict, name)
            # push survived value to set matching the name 
            push!(get(family_dict, name, nothing), value.Survived)
        else
            # if dict does not contain this family, create a new set containing the "Survived" value of this person
            family_dict[name] = [value.Survived]
        end
    end
    
    return family_dict
end
```

### We get the best score when predicting that all women survive but also taking into consideration the surival of the whole family.
Let's now take a look at the misclassifications and try to improve our approach.

```julia
function score_confusion(data, predicted, actual)
    println(typeof(data))
    println(data[3, :Sex] == "male")
    
    male_fd=0; male_fs=0; female_fd=0; female_fs=0
    
    correct = 0
    for i in 1:size(predicted)[1]
        if predicted[i] == actual[i]
            correct += 1
        else
            if data[i, :Sex] == "male"
                if predicted[i] == 1
                    male_fs += 1
                else
                    male_fd +=1
                end
            else
                if predicted[i] == 1
                    female_fs += 1
                else 
                    female_fd += 1
                end
            end
        end
    end
    println("Survived males predicted to die: $male_fd")
    println("Deceased males predicted to survive: $male_fs")
    
    println("Survived females predicted to die: $female_fd")
    println("Deceased females predicted to survive: $female_fs")
    return correct / size(predicted)[1]
end
```

```julia
train, test = partition(eachindex(y), 0.8, shuffle=true)
family_dict = get_family_dict(data[train, :])
ŷ = get_prediction(X[test, :], predict_all_women_survive_using_family, family_dict)

score_confusion(X[test, :], ŷ, y[test])
```
